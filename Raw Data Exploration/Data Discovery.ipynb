### Convert the `Date Joined` column to datetime

Next, I'll use the `to_datetime()` function from the `pandas` library  to convert the `Date Joined` column to datetime. This splits each value into year, month, and date components. This is an important step in data cleaning, as it makes the data in this column easier to use in tasks you may encounter.

```python
# Step 1: Use pd.to_datetime() to convert Date Joined column to datetime
# Step 2: Update the column with the converted values

companies['Date Joined'] = pd.to_datetime(companies['Date Joined'])
```

```python
# Use .info() to confirm that the update actually took place

companies.info()
```

### Create a `Year Joined` column

It will be common to encounter situations where I will need to compare the year joined with the year founded. The `Date Joined` column does not just have year—it has the year, month, and date.

```python
# Step 1: Use .dt.year to extract year component from Date Joined column
# Step 2: Add the result as a new column named Year Joined to the DataFrame

companies['Year Joined'] = companies['Date Joined'].dt.year
```

```python
# Use .head() to confirm that the new column did get added

companies.head()
```

## Step 4: Results and evaluation


### Take a sample of the data

Although it's not necessary for this analysis, I'll take a sample of the data to practice for future scenarios where I might need to due to time and resource limitations. I'll use the `sample()` function with a specified random_state for reproducibility.

I will use `sample()` with the `n` parameter set to `50` to randomly sample 50 unicorn companies from the data. Specifying the `random_state` parameter will ensure reproducibility of my work.

```python
# Sample the data

companies_sampled = companies.sample(n = 50, random_state = 42)

companies_sampled
```

### Visualize the time it took companies to reach unicorn status

To visualize the longest time it took companies to reach unicorn status for each industry in the sample, I will create a bar plot using the `bar()` function from the `matplotlib.pyplot` module. First, I'll need to prepare the data.

```python
# Prepare data for plotting

companies_sampled['Years Till Unicorn'] = companies_sampled['Year Joined'] - companies_sampled['Year Founded']

grouped_companies_sampled = (companies_sampled[['Industry','Years Till Unicorn']]
                             .groupby('Industry')
                             .max()
                             .sort_values(by = ('Years Till Unicorn'))
                            )

grouped_companies_sampled2 = (companies_sampled[['Industry','Years Till Unicorn']]
                             .groupby('Industry')
                             .agg(['max', 'mean'])
                             .sort_values(by = ('Years Till Unicorn', 'max'))
                            )

grouped_companies_sampled
```

```python
# Creating the bar plot
# with the various industries as the categories of the bars
# and the time it took to reach unicorn status as the height of the bars

plt.bar(x = grouped_companies_sampled.index, height = grouped_companies_sampled['Years Till Unicorn'])

# Set title

plt.title('Maximum years taken to become unicorn per industry from sample')

# Set x-axis label

plt.xlabel('Industry')

# Set y-axis label
plt.ylabel('Years')

# Rotate labels on the x-axis as a way to avoid overlap in the positions of the text

plt.xticks(rotation=40, horizontalalignment = 'right')

# Display the plot

plt.show()
```

**Question:** Any noteworthy observations?

- This bar plot shows that for this sample of unicorn companies, the largest value for maximum time taken to become a unicorn occurred in the Heath and Fintech industries, while the smallest value occurred in the Consumer & Retail industry.

### Visualize the maximum unicorn company valuation per industry

I will create a visualization of unicorn companies' maximum valuation for each industry represented in the sample. To create a bar plot to visualize this, I'll use the `bar()` function from the `matplotlib.pyplot` module. Before plotting, I'll need to create a new column that represents the companies' valuations as numbers (instead of strings, as they're currently represented).

```python
# Create a column representing company valuation as numeric data
companies_sampled['Valuation in Billions'] = companies_sampled['Valuation']

# Remove '$'
companies_sampled['Valuation in Billions'] = companies_sampled['Valuation in Billions'].str.replace('$', '')

# Remove 'B'
companies_sampled['Valuation in Billions'] = companies_sampled['Valuation in Billions'].str.replace('B', '')

# Convert to int
companies_sampled['Valuation in Billions'] = companies_sampled['Valuation in Billions'].astype(int)

companies_sampled.head()

```

```python
# Prepare data

grouped_companies_sampled = (companies_sampled[['Industry', 'Valuation in Billions']]
                             .groupby(['Industry'])
                             .max()
                             .sort_values(by=['Valuation in Billions'])
                            )

grouped_companies_sampled
```

```python
# Create bar plot
# with the various industries as the categories of the bars
# and the maximum valuation for each industry as the height of the bars


plt.bar(x = grouped_companies_sampled.index, height = grouped_companies_sampled['Valuation in Billions'])

# Set title

plt.title('Maximum valution per industry from sample')

# Set x-axis label

plt.xlabel('Industry')

# Set y-axis label

plt.ylabel('Max Valuation')

# Rotate labels on the x-axis as a way to avoid overlap in the positions of the text

plt.xticks(rotation=40, horizontalalignment = 'right')


# Display the plot

plt.show()
```

**Question:** What did I observe from this bar plot?

- The bar plot reveals that in the sampled data, Artificial Intelligence companies are reaching the highest valuations as unicorns. In contrast, industries like Auto & Transportation and Consumer & Retail are seeing the lowest peak valuations among their unicorn companies.
